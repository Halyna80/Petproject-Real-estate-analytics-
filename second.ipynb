```python
# Імпортування бібліотеки
import pandas as pd
```

```python
# імпортування даних в датафрейм
secondy = pd.read_csv('apartment_second_city.csv', delimiter=';', encoding='cp1251')
print(secondy)
```

```python
# Розрахунок середньої ціни за квадратний метр для кожного міста
average_price_by_city = secondy.groupby('city_name')['average_price_per_square_meter'].mean().reset_index()

# Виведення даних
print(average_price_by_city)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Коробкова діаграма (Box Plot)
plt.figure(figsize=(10, 6))
sns.boxplot(x='average_price_per_square_meter', data=average_price_by_city, color='skyblue')
plt.title('Середня вартість кв м по містам України', fontsize=15)
plt.xlabel('Ціна за кв м', fontsize=12)
plt.ylabel('Міста', fontsize=12)
plt.savefig('boxplot_secondy.png')
plt.show()
```
